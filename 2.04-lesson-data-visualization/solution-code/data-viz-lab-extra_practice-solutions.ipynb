Remember the six "lessons" of visualizing data.

- Lesson 1: Understand the context. (Who, What, How)
- Lesson 2: Choose an appropriate visual display. (What type of graph?)
- Lesson 3: Eliminate clutter. (What can I delete or make more subtle?)
- Lesson 4: Focus attention where you want it. (How do I emphasize important things?)
- Lesson 5: Think like a designer. (How do I organize my visualization?)
- Lesson 6: Tell a story! (How do I communicate my visualization?)

Today, we're going to walk through `matplotlib`, create some graphs, and tie these in with the principles we've learned.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

First, let's randomly generate some data so that we can plot it!

```python
np.random.seed(42)
x_1 = np.random.normal(loc = 100, scale = 15, size = 100)
x_2 = np.random.normal(loc = 110, scale = 5, size = 100)
y = x_1 + x_2 + np.random.normal(loc = 0, scale = 25, size = 100)

df = pd.DataFrame().from_dict({'y':y,
                              'x1':x_1,
                              'x2':x_2})
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
```

<details><summary>Let's visualize the distribution of $X_1$. What type of graph is probably best here?</summary>
Histogram (a vertical bar chart, but for quantitative variables).
</details>

```python
plt.hist(x_1)
```

```python
plt.hist(df['x1']);
```

Let's change the color.

```python
plt.hist(df['x1'], bins = 10, color='g');
```

```python
plt.hist(df['x1'], bins = 10, color='g', alpha = 0.1);
```

Let's add some labels!

```python
plt.hist(df['x1'], bins = 10, color='g', alpha = 0.1);
plt.xlabel("IQ")
plt.ylabel("Frequency");
```

Let's move the labels to a better spot.

```python
plt.hist(df['x1'], bins = 10, color='g', alpha = 0.1);
plt.xlabel("IQ", position = (0,0)) # (x,y)
plt.ylabel("Frequency", position = (0, 1)); # (x,y)
```

Can we automatically move these labels so that they're aligned with the edge of the axis?

```python
plt.hist(df['x1'], bins = 10, color='g', alpha = 0.1);
plt.xlabel("IQ", position = (0,0), ha = 'left') # horizontal alignment
plt.ylabel("Frequency", position = (0, 1), ha = 'right'); # horizontal alignment
```

```python
plt.hist(df['x1'], bins = 10, color='g', alpha = 0.1);
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey'); # (x,y)
```

Let's compare the two distributions on the same histogram.

```python
plt.hist(df['x1'], bins = 10, color='g', alpha = 0.5)
plt.hist(df['x2'], bins = 10, color='g', alpha = 0.5)
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey'); # (x,y)
```

```python
plt.hist(df['x1'], bins = 10, color='g', alpha = 0.5)
plt.hist(df['x2'], bins = 10, color='r', alpha = 0.5)
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey'); # (x,y)
```

```python
plt.hist(df['x1'], bins = 10, color='b', alpha = 0.5)
plt.hist(df['x2'], bins = 10, color='orange', alpha = 0.5)
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey') # (x,y)
plt.title("Comparing IQs");
```

```python
plt.hist(df['x1'], bins = 10, color='b', alpha = 0.5)
plt.hist(df['x2'], bins = 10, color='orange', alpha = 0.5)
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey') # (x,y)
plt.title("Comparing IQs", position = (0,1), ha = 'left');
```

Let's add a legend.

```python
plt.hist(df['x1'], bins = 10, color='b', alpha = 0.5, label = 'all')
plt.hist(df['x2'], bins = 10, color='orange', alpha = 0.5, label = 'DSI')
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey') # (x,y)
plt.legend()
plt.title("Comparing IQs", position = (0,1), ha = 'left');
```

Let's give it a better title.

```python
plt.hist(df['x1'], bins = 10, color='b', alpha = 0.5, label = 'all')
plt.hist(df['x2'], bins = 10, color='orange', alpha = 0.5, label = 'DSI')
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey') # (x,y)
plt.title("The IQ of DSI students is, on average, higher than the population.", position = (0,1), ha = 'left');
```

Let's fix that overhang!

```python
plt.hist(df['x1'], bins = 10, color='b', alpha = 0.5, label = 'all')
plt.hist(df['x2'], bins = 10, color='orange', alpha = 0.5, label = 'DSI')
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey') # (x,y)
plt.title("The IQ of DSI students is, on average,\nhigher than the population.", position = (0,1), ha = 'left');
```

```python
plt.hist(df['x1'], bins = 10, color='b', alpha = 0.5, label = 'all')
plt.hist(df['x2'], bins = 10, color='orange', alpha = 0.5, label = 'DSI')
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 1), ha = 'right', color = 'grey') # (x,y)
plt.title("The IQ of DSI students is, on average,\nhigher than the population.", position = (0,1), ha = 'left', fontsize=16);
```

Let's break this into two plots.

```python
fig, ax = plt.subplots(2)
ax[0].hist(df['x1'], bins = 10, color='b', alpha = 0.5, label = 'all')
ax[1].hist(df['x2'], bins = 10, color='orange', alpha = 0.5, label = 'DSI')
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 2), ha = 'right', color = 'grey') # (x,y)
plt.suptitle("The IQ of DSI students is, on average,\nhigher than the population.", position = (0,1), ha = 'left', fontsize=16, va = 'top');
#plt.figure(figsize = (30,20));
```

```python
fig, ax = plt.subplots(2, sharex=True, sharey=True)
ax[0].hist(df['x1'], bins = 10, color='b', alpha = 0.5, label = 'all')
ax[1].hist(df['x2'], bins = 10, color='orange', alpha = 0.5, label = 'DSI')
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 2), ha = 'right', color = 'grey') # (x,y)
plt.suptitle("The IQ of DSI students is, on average,\nhigher than the population.", position = (0,1), ha = 'left', fontsize=16, va = 'top');
```

```python
fig, ax = plt.subplots(2, sharex=True, sharey=True, figsize = (10,10))
ax[0].hist(df['x1'], bins = 10, color='b', alpha = 0.5, label = 'all')
ax[1].hist(df['x2'], bins = 10, color='orange', alpha = 0.5, label = 'DSI')
plt.xlabel("IQ", position = (0,0), ha = 'left', color = 'grey') # (x,y)
plt.ylabel("Frequency", position = (0, 2), ha = 'right', color = 'grey') # (x,y)
plt.suptitle("The IQ of DSI students is, on average,\nhigher than the population.", position = (0,1), ha = 'left', fontsize=16, va = 'top');
```
